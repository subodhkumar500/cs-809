```
Data=data.drop("DATE",axis=1)
print(Data.shape)
```

```
#defining covarianve matrix
covMatrix = pd.DataFrame.cov(data)
print (covMatrix)
```

```
corrMatrix = data.corr() #defining correlation matrix
print (corrMatrix)
print ( "The shape of variancecorrelation matrix = ", corrMatrix.shape)
```

```
#computing eigen values and eigen vector using Lingalg.eig()
eig_values, eig_vectors = np.linalg.eig(corrMatrix) 
print(eig_values)
```

```
print('Eigenvectors \n%s' %eig_vectors)
```

```
#sorting eigen values
eig_values = np.sort(eig_values)
print(eig_values)
```

```
(100*(eig_values[3] / sum(eig_values)))+(100*(eig_values[4] / sum(eig_values)))+(100*(eig_values[5] / sum(eig_values)))
```

```
eig_vectors.T[0]
```

```
eig_vectors.T[4]
```

```
eig_vectors.T[5]
```

```
Projected_X = Data.dot(eig_vectors.T[0])
```

```
Projected_y = Data.dot(eig_vectors.T[4])
```

```
Projected_z = Data.dot(eig_vectors.T[5])
```

```
result=pd.DataFrame(Projected_X,columns=['PC1'])
result['PC2']=Projected_y
result['PC3']=Projected_z
dataframe=result
print(dataframe.head(2132))
```
